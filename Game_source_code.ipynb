```python
import random
import cv2
import cvzone
from cvzone.HandTrackingModule import HandDetector
import time
 
cap = cv2.VideoCapture(0)
# setting video capture to a same value so that it doesn't change when the device changes
# 3 = width
cap.set(3, 640) 
# 4 = height
cap.set(4, 480)
 
# detects only one hand 
detector = HandDetector(maxHands=1)
 
timer = 0
stateResult = False
startGame = False
scores = [0, 0]  # [AI, Player]
 
while True:

    # if the bgimg was outside the loop the images will remain on the bg img every time it iterates. to avoid that and have a new bg every time use it inside while loop
    imgBG = cv2.imread("Resources/BG.png")
    cv2.putText(imgBG, str('Press \'s\' to'), (505, 180), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 255), 3)
    cv2.putText(imgBG, str('Start'), (580, 220), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 255), 3)
    cv2.putText(imgBG, str('Press \'q\' to'), (505, 620), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 0), 3)
    cv2.putText(imgBG, str('Quit'), (590, 680), cv2.FONT_HERSHEY_PLAIN, 3, (0, 0, 0), 3)
    success, img = cap.read()
 
    # img to be resized, specific pixel number, output img, 420/480 = sqeezing down to the height wanted
    imgScaled = cv2.resize(img, (0, 0), None, 0.875, 0.875)
    # cropping 640-400 = 240/2 = 120
    imgScaled = imgScaled[:, 120:520]
 
    # Find Hands
    hands, img = detector.findHands(imgScaled)  # with draw
 
    if startGame:
 
        if stateResult is False:
            timer = time.time() - initialTime
            # cv2.putText(imgBG, str(''), (605, 435), cv2.FONT_HERSHEY_PLAIN, 4, (255, 0, 255), 4)
 
            if timer > 0.5:
                stateResult = True
                timer = 0
 
                if hands:
                    playerMove = None
                    # first hand detected
                    hand = hands[0]

                    fingers = detector.fingersUp(hand)

                    print(fingers)

                    if fingers == [0, 0, 0, 0, 0]:
                        playerMove = 1
                        cv2.putText(imgBG, str('Rock'), (540, 435), cv2.FONT_HERSHEY_PLAIN, 4, (0, 0, 0), 6)
                    elif fingers == [1, 1, 1, 1, 1]:
                        playerMove = 2
                        cv2.putText(imgBG, str('Paper'), (540, 435), cv2.FONT_HERSHEY_PLAIN, 4, (0, 0, 0), 6)
                    elif fingers == [0, 1, 1, 0, 0]:
                        playerMove = 3
                        cv2.putText(imgBG, str('Scissor'), (540, 435), cv2.FONT_HERSHEY_PLAIN, 4, (0, 0, 0), 6)
                    else:
                        cv2.putText(imgBG, str('Wrong'), (540, 435), cv2.FONT_HERSHEY_PLAIN, 4, (0, 0, 0), 6)
 
                    randomNumber = random.randint(1, 3)
                    imgAI = cv2.imread(f'images/{randomNumber}.png', cv2.IMREAD_UNCHANGED)
                    # imgAI = cv2.imread('images/s.png',cv2.IMREAD_UNCHANGED)
                    imgBG = cvzone.overlayPNG(imgBG, imgAI, (149, 310))
 
                    # Player Wins
                    if (playerMove == 1 and randomNumber == 3) or \
                            (playerMove == 2 and randomNumber == 1) or \
                            (playerMove == 3 and randomNumber == 2):
                        scores[1] += 1
 
                    # AI Wins
                    if (playerMove == 3 and randomNumber == 1) or \
                            (playerMove == 1 and randomNumber == 2) or \
                            (playerMove == 2 and randomNumber == 3):
                        scores[0] += 1
 
    # overlapping
    # height: 654-234=420 width:1195-795=400 
    imgBG[234:654, 795:1195] = imgScaled
 
    if stateResult:
        imgBG = cvzone.overlayPNG(imgBG, imgAI, (149, 310))
 
    cv2.putText(imgBG, str(scores[0]), (410, 215), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 6)
    cv2.putText(imgBG, str(scores[1]), (1112, 215), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 6)
 
    # cv2.imshow("Image", img)
    cv2.imshow("Rock Paper Scissor", imgBG)
    # cv2.imshow("Scaled", imgScaled)
 
    key = cv2.waitKey(1)
    if key == ord('s'):
        startGame = True
        initialTime = time.time()
        stateResult = False

    elif key & 0xFF == ord('q'):
        break
        
cap.release() 
cv2.destroyAllWindows()
```
